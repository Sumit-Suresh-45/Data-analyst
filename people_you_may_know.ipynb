```python
import json
def load_data(filename):
    with open(filename,'r') as f:
        return json.load(f)
def suggestions(data,user_id):
    user_friends={} #stotres the friends to corresponding id
    for user in data["users"]:
        user_friends[user['id']]=set(user['friends'])
    if user_id not in user_friends:
        return[]
    direct_friends=user_friends[user_id]
    suggestions={}
    for  friends in direct_friends:
        for mutual in user_friends[friends]:
            if mutual != user_id and mutual not in direct_friends:
                suggestions[mutual]=suggestions.get(mutual,0)+1

    sorted_suggestions=sorted(suggestions.items(),key = lambda x:x[1] ,reverse=True)
    return [user for user ,_ in sorted_suggestions]
data=load_data("massive.json")
user_id = 1
recc=suggestions(data,user_id)
print(f"People You May Know For USER ID {user_id} - {recc}")
```
