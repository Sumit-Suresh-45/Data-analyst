```python
import json
```

```python
def data_load(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
def data_display(data):
    print("User's Connections and Friends: ")
    for user in data["users"]:
        print(f"ID: {user['id']} - {user['name']} - Friends: {user['friends']} - Liked Pages: {user['liked_pages']}")
    print("\nData of Related Pages: ")
    for page in data["pages"]:
        print(f"{page['id']} - {page['name']}")
```

```python
data=data_load("data_dump.json")
```

```python
data_display(data)
```
